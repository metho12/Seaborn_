```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
cl = ['black','red','hotpink','yellow','pink']
ex = [.2,.3,.4,.5,.6]

plt.pie(x,labels=t,explode=ex,colors=cl)
plt.legend(title="Categories")
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
cl = ['black','red','hotpink','yellow','pink']
ex = [.2,.3,.4,.5,.6]

sns.set()

plt.pie(x,labels=t,explode=ex,shadow = True ,colors=cl)
plt.legend(title="Categories")
```

```python
# cauntplot
sns.countplot(x='embark_town',data=df)
```

```python
sns.countplot(x='survived',data=df)
```

```python
sns.countplot(x='survived',hue= 'sex',data=df)
```

```python
sns.countplot(x='survived',hue= 'sex',data=df,palette='gist_gray')
```

```python
# barplot

sns.barplot(x='sex',y= 'survived',data=df)
```

```python
# barplot

sns.barplot(x='pclass',y= 'survived',data=df)
```

```python
# barplot

sns.barplot(x='pclass',y= 'survived',hue='sex',data=df)
```

```python
# barplot

sns.barplot(x='pclass',y= 'survived',hue='sex',data=df,ci=None)
```

```python
# barplot

sns.barplot(x='pclass',y= 'survived',hue='sex',data=df,color='black')
```

```python
# barplot

sns.barplot(x='c',y= 'survived',hue='sex',data=df,estimator=sum)
```

```python
# pie chart
import numpy as np
import matplotlib.pyplot as plt


x = np.array([10,20,30,40,50])

plt.pie(x)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']

plt.pie(x,labels=t)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']

plt.pie(x,labels=t,startangle=90)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
ex = [.2,.3,.4,.5,.6]
plt.pie(x,labels=t,explode=ex)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
ex = [.2,.3,.4,.5,.6]
plt.pie(x,labels=t,explode=ex,shadow=True)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
cl = ['black','red','hotpink','yellow','pink']
ex = [.2,.3,.4,.5,.6]
plt.pie(x,labels=t,explode=ex,colors=cl)
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
cl = ['black','red','hotpink','yellow','pink']
ex = [.2,.3,.4,.5,.6]
plt.pie(x,labels=t,explode=ex,colors=cl)
plt.legend()
```

```python
x = np.array([10,20,30,40,50])
t = ['web development','Data science','mobile','Deep learning','machine learning']
cl = ['black','red','hotpink','yellow','pink']
ex = [.2,.3,.4,.5,.6]
plt.pie(x,labels=t,explode=ex,colors=cl)
plt.legend(title="Categories")
```
